```python
# ライブラリの読み込み
import MeCab
from zodbpickle import pickle
from collections import OrderedDict
```

```python
# テキストの読み込み
f = open('haiku-vocab-0.txt')
data = f.read()  # ファイル終端まで全て読んだデータを返す
f.close()
```

```python
data = data.split('\n')
```

```python
data
```

```python
# 本来分かち書きするための俳句のデータ
len(data)
```

```python
# 読み込んだテキストの分かち書き
m = MeCab.Tagger("-Owakati")
```

```python
temp_data = data[:3]
```

```python
temp_data
```

```python
# 俳句に出てきた単語を追加する
word_list = list()
for s in temp_data:
    
    word_list.extend(m.parse(s).replace('\n', '').split(' '))
```

```python
word_list
```

```python
# 俳句に出てきた単語を追加
word_list = list()
for s in data:
    
    word_list.extend(m.parse(s).replace('\n', '').split(' '))
```

```python
# 単語の種類
len(word_list)
```

```python
# 分かち書きされた単語のリスト
word_list
```

```python
# 単語の重複を削除
word_list = list(set(word_list))
```

```python
word_list
```

```python
# 順番が変わらない辞書の作成
vocab = OrderedDict()
```

```python
# 辞書
vocab
```

```python
# 単語に対して辞書を作成
for s in word_list:
    
    vocab[s] = 0
```

```python
# 作成された辞書
vocab
```

```python
# 辞書の保存
pickle.dump(vocab, open('haiku-vocab-0.pkl','wb'), protocol=3 )
```

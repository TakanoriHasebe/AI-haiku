```python
# ライブラリのロード
from zodbpickle import pickle
import codecs
import MeCab
import os
import glob
from collections import OrderedDict
import pickle
```

```python
# 辞書の読み込み
with open('../vocab/haiku-vocab-0.pkl', 'rb')\
     as f: vocab = pickle.Unpickler(f).load()
```

```python
# 辞書
vocab
```

```python
# 分かち書き
m = MeCab.Tagger("-Owakati")
```

```python
# bag of wordsを作成する変数
bag_of_words = []
```

```python
# 芭蕉のテキストの読み込み
fin = codecs.open('../basho-haiku-data/basho_haiku_list.txt', 'r', 'utf-8')
txtRaw = fin.read()
fin.close()
```

```python
# 読み込んだ現代俳句をリストに変換
basho_haiku_list = txtRaw.split('\n')
```

```python
# 俳句の表示
basho_haiku_list
```

```python
# まずは、簡単な文章に対して、辞書を用いてbag_of_wordsを作成
temp_list = basho_haiku_list[:2]
```

```python
temp_list
```

```python
# 俳句を分かち書き
temp_wakati_haiku = []
for s in temp_list:
    
    temp_wakati_haiku.append(m.parse(s).replace(' \n', '').split(' '))
print(temp_wakati_haiku)  
```

```python
# 簡単な文章に対してbag-of-wordsを作成
for i in range(0, len(temp_wakati_haiku)):
    
    # 一文字づつ見て、辞書に出現回数を追加
    for s in temp_wakati_haiku[i]:
        
        # print(s)
        vocab[s] += 1
    
    # bag_of_wordsの作成
    bag_of_words.append(list(vocab.values()))
    
    # vocabを初期化
    for s in vocab:
        vocab[s] = 0
```
